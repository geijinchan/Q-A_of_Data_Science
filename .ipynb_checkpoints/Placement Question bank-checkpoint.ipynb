# Q2
- Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
  he can remove just one character at the index in the string, and the remaining characters will occur the same
  number of times. Given a string, determine if it is valid. If so, return *YES* , otherwise return *NO* .

```python
# I have used list comprehension here
text = input('Enter the string :')
#items = []
#for i in text:
#    items.append(i)
items = [i for i in text]
dic = {} 
dic = {i: items.count(i) for i in items}
occ = list(dic.values())
max_occ = max(occ)
max_count = occ.count(max_occ)
if len(set(occ)) == 1 or (max_count == 1 and (max_count-1) * (len(occ) - 1) + 1 == sum(occ)):
    print("YES")
else:
    print("NO")
```

# Q3

```python
import pandas as pd
import requests
url = 'https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json'
try:
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data['pokemon'])
    df = df.dropna()
    
    # Changing data type
    df['num'] = df['num'].astype(int)
    df['height'] = df['height'].str.replace('m','').astype(float)   # removing m from the data and converting to float
    df['weight'] = df['weight'].str.replace('kg','').astype(float)  # same as above
    df['candy_count'] = df['candy_count'].astype(int)
    df['spawn_chance'] = df['spawn_chance'].astype(float)
    
    df.to_excel('data.xlsx', index = False)
    
except requests.exceptions.RequestException as e:
    print(f'Error :{e}')
```

```python
import requests
import json
import pandas as pd
import re                             # for regular expression for summary column/features
from datetime import datetime

url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"
response = requests.get(url)
data = response.json()

records = []
for episode in data["_embedded"]["episodes"]:
    record = {
        "id": episode["id"],
        "url": episode["url"],
        "name": episode["name"],
        "season": episode["season"],
        "number": episode["number"],
        "type": episode["type"],
        "airdate": episode["airdate"],
        "airtime": episode["airtime"],
        "runtime": episode["runtime"],
        "average_rating": episode["rating"]["average"],
        "summary": re.sub('<.*?>', '', episode['summary']),
        "medium_image_link": episode["image"]["medium"],
        "original_image_link": episode["image"]["original"]
    }
    records.append(record)
df = pd.DataFrame(records)     # Converting list into datafram
df1 = df.copy()
df.to_csv("CSV_data.csv", index=False)       # Saving data in CSV file
df1['airdate'] = pd.to_datetime(df1['airdate'])
df1.dtypes
```

# Q 10)

Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.
Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

```python
# A program that can count number of verbs, nouns, pronouns and adjectives is a passage or sentences and count it 
# for that NLP is to used a library name nltk
```

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize
nltk.download('averaged_perceptron_tagger')
```

```python
verb = []
noun = []
adjective = []
pronoun = []
a = 'This is abhishek and he is studying M.Sc.'
b = word_tokenize(a)                               # Creating token 
c = pos_tag(b)                                     # part of speech tagging 
```

```python
for word, pos in c:
    if pos.startswith('VB'):
        verb.append(pos)
    elif pos.startswith('NN'):
        noun.append(pos)
    elif pos == 'PRP' or pos == 'PRP$':
        pronoun.append(pos)
    elif pos.startswith('JJ'):
        adjective.append(pos)
```

```python
print(f'Total noun {len(noun)},\nTotal verb {len(verb)},\nTotal adjective {len(adjective)},\nTotal Pronoun {len(pronoun)}')
```

# Q2
- Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
  he can remove just one character at the index in the string, and the remaining characters will occur the same
  number of times. Given a string, determine if it is valid. If so, return *YES* , otherwise return *NO* .

```python
# I have used list comprehension here
text = input('Enter the string :')
#items = []
#for i in text:
#    items.append(i)
items = [i for i in text]
dic = {} 
dic = {i: items.count(i) for i in items}
occ = list(dic.values())
max_occ = max(occ)
max_count = occ.count(max_occ)
if len(set(occ)) == 1 or (max_count == 1 and (max_count-1) * (len(occ) - 1) + 1 == sum(occ)):
    print("YES")
else:
    print("NO")
```

```python
# Importing important packages
import requests
import pandas as pd
link = 'https://data.nasa.gov/resource/y77d-th95.json'
response = requests.get(link)
data = response.json()
```

```python
df = pd.DataFrame(data)     # Didn't created a copy of data set just becuase data vairable is storing the data already
```

### Removing NA columns

```python
# removing nan columns
df.drop(columns=[':@computed_region_cbhk_fwbd', ':@computed_region_nnqa_25f4'],inplace=True)
```

```python
df.head()
```

### Changing column name

```python
df.rename(columns={'name':'Name of Earth Meteorite',
                   'id':'ID of Earth Meteorite', 
                   'mass':'Mass of Earth Meteorite',
                   'year':'Year at which Earth Meteorite was hit',
                  },inplace=True)
df
```

### Working on data types

##### String data types
- As string data type is already in object types

```python
# Before changing data types of columns making sure to deal with the null values
for col in df.columns:
    if df[col].isnull().sum() > 0:
        print(col)
```

```python
column = [col for col in df.columns if df[col].isnull().sum() > 0]
```

```python
df['Mass of Earth Meteorite'] = df['Mass of Earth Meteorite'].astype(float)
df['ID of Earth Meteorite'] = df['ID of Earth Meteorite'].astype(int)
df['reclat'] = df['reclat'].astype(float)
df['reclong'] = df['reclong'].astype(float)
```

```python
df['Year at which Earth Meteorite was hit'] = df['Year at which Earth Meteorite was hit'].str.split('T').str.get(0)
df['Year at which Earth Meteorite was hit'] = pd.to_datetime(df['Year at which Earth Meteorite was hit'],errors='coerce')
```

```python
df = df.dropna()
```

```python
df['geolocation'] = df['geolocation'].apply(lambda coord: [int(c) for c in coord['coordinates']] if isinstance(coord, dict) and 'coordinates' in coord else None)
df = df.dropna(subset=['geolocation'])
df = df.reset_index(drop=True)
```
